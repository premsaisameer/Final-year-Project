```python
import pandas as pd
import numpy as np 
import seaborn as sns
from matplotlib import pyplot as plt
import sklearn
```

```python
data = pd.read_excel("datasets/McDonald_menu.xlsx")
```

```python
data.info()
```

```python
data["Category"].value_counts().plot.pie()
```

```python
# Shows if any of the columns have missing values
data
```

```python
print(data["Category"].unique() )
```

```python
#Breakfast

breakfast_menu = data[data["Category"] == 'Breakfast']
```

```python
#Main menu

main_menu = data[(data["Category"] == 'Chicken & Fish') | (data["Category"] == 'Beef & Pork') 
                 |(data["Category"] == 'Salads')]
```

```python
#Bevarages or drinks 

bevarages_menu = data[(data["Category"] == 'Coffee & Tea') | (data["Category"] == 'Smoothies & Shakes') 
               | (data["Category"] == 'Beverages')]
```

```python
#Extras or Side dishes or snacks
extras_menu = data[(data["Category"] == 'Desserts') | (data["Category"] == 'Snacks & Sides')]
```

```python
measures = ['Calories', 'Total Fat', 'Cholesterol','Sodium', 'Sugars', 'Carbohydrates']
for m in measures:   
    plot = sns.swarmplot(x="Category", y=m, data=data)
    plt.setp(plot.get_xticklabels(), rotation=45)
    plt.title(m)
    plt.show()
    plt.savefig(m)
```

```python
# Identifies the top 5 highest calorie menu items
data.nlargest(5, 'Calories')
```

This is quite a suprising result because I was not expecting so many breakfast menu items to be so high in calories compared to the lunch menu.

```python
# Identifies the top 5 lowest calorie menu items
data.nsmallest(5, 'Calories')
```

Now, I was expecting the lowest calorie menu items to be drinks because most diet drinks are zero calorie.

<h2> Unhealthiest & Healthiest side menu items </h2>

```python
sides_plot = sns.stripplot(x='Category', y='Calories', data=extras_menu, size=7)
# assigns labels to the x axis and sets them to a 45 degree angle
sides_plot.set_xticklabels(sides_plot.get_xticklabels(), rotation= 45)
plt.title('McDonald\'s sides menu items by calories')
plt.show()
```
